```python
# converting amount to cr. rupees
# as in amount have some missing values so converting it into float
# filling missing value with "0" it means "undislocsed" not actual zero
```

```python
df['amount'] = df['amount'].fillna('0')
```

```python
# dtype of amount is still "object" now converting it to "integer"
```

```python
#amount.astype('int')
# this function will give error as all amount value given with comma. Ex - 15,00,000
# so firstly I will remove comma.
```

```python
# using str function as amount still in object dtype. and again storing in df.
#df.amount = df.amount.str.replace(',','')
```

```python
# as on some places still there is "undisclosed amount" so converting in "integer" will give "value error"
# Value Error: Invalid  literal for int() with base10: 'undisclosed'
```

```python
# replacing 'undisclosed with 0' in amount
df['amount'] = df['amount'].str.replace(',','')
df['amount'] = df['amount'].str.replace('undisclosed','0')
df['amount'] = df['amount'].str.replace('unknown','0')
df['amount'] = df['amount'].str.replace('Undisclosed','0')
```

```python
df = df[df['amount'].str.isdigit()]
```

```python
df['amount'] = df['amount'].astype('float')
```

```python
df.info()
```

```python
df.head()
```

```python
# Now converting amount in Cr.Rupees
```

```python
def to_inr(dollar):
    inr = dollar * 82.5
    return inr/10000000
```

```python
# checking function with random value
to_inr(3000000.0)
```

```python
# Now applying "to_inr" function into "amount"column.

df['amount'] = df['amount'].apply(to_inr)
```

```python
#checking data again
df.head()
```

```python
# converting date in right format firstly by replacing
df['date'] = df['date'].str.replace('05/072018','05/07/2018')
```

```python
# Now replacing "Datecolumn" in "Date Time " object.and also in errors place using "coerce"
# it will remove those value which is non convertible in date time format.
df['date'] = pd.to_datetime(df['date'], errors = 'coerce')
```

```python
# Removing rows with missing values with all columns.
# doing dropna for all column removing missing value except 'subvertical' and if missing value is
# less than 10per of data will remove them.
```

```python
df = df.dropna(subset=['date','startup','vertical','city','investors','round','amount'])
```
