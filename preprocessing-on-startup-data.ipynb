#### Renaming the column name

```python
df.rename(columns = {
    'Date dd/mm/yyyy': 'date',
    'Startup Name': 'startup',
    'Industry Vertical': 'vertical',
    'SubVertical': 'subvertical',
    'City  Location':  'city',
    'Investors Name':'investors',
    'InvestmentnType': 'round',
    'Amount in USD': 'amount'
}, inplace= True)
```

```python
df.head()
```

#### Convert Amount into Crore Rupees

- As there is Missing value in "amount" only out of 3043 - 2084 rows have amount data. 
- So firstly adding zero on na place

```python
df['amount'] = df['amount'].fillna('0') 
```

```python
df.info()
```

- Now there is no missing value in "amount column".
- Now can apply astype function to change amount type from "object to integer"

```python
df['amount'].astype('int')
```

- We are getting "ValueError" as in amount data there is "comma" so we need to remove comma from all amount.

```python
df['Startup Name'].unique()
```

```python
df['amount'] = df['amount'].str.replace(',','')
```

```python
df['amount']
```

- Now again trying to convert datatype into Integer after removing comma.

```python
df['amount'].astype('int')
```

- again getting error "ValueError: invalid literal for int() with base 10: 'undisclosed'" so will replace "undisclosed value with zero"

```python
df['amount'] = df['amount'].str.replace('undisclosed','0')
```

```python
df['amount'].astype('int')
```

- also converting this given error "ValueError: invalid literal for int() with base 10: 'unknown'" into zero.

```python
df['amount'] = df['amount'].str.replace('unknown','0')
```

```python
df['amount'].astype('int')
```

- Now get error "Undisclosed" with capital "U" so will change again with "zero"

```python
df['amount'] = df['amount'].str.replace('Undisclosed','0')
```

```python
df['amount'].astype('int')
```

- Next error get "'4889975.54' so will choose amount only when its in digit only.

```python
df = df[df['amount'].str.isdigit()]
```

```python
df['amount'].astype('float')
```

```python
df['amount'] = df['amount'].astype('float')
```

```python
df.info()
```

- Now can see 'Amount' column converted into 'float64' with '3029' rows.

```python
df.head(2)
```

- Now after this much cange in amount finally will convert this into "crore rupees"

```python
def to_inr(dollar):
    inr = dollar * 82.5
    return inr/10000000
    
```

```python
df['amount'] =  df['amount'].apply(to_inr)
```

```python
df['amount']
```

```python
df.head(2)
```
